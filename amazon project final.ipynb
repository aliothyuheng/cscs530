#Customer Purchase Behavior Model
##Introduction
    1.Background
    Buying products online is becoming more and more popular these days since its convenience and lower cost. However, the disadvantage of online shopping is that the customer needs to spend more time finding the products. To solve this problem, online retailers are implementing a recommendation system to help the customers find what they are looking for. It learns from the customer behaviors and give predictions also based on the customer behaviors. 
    
    There are two major algorithms widely used in the industry world: collaborative filtering and cluster models. These two algorithms were reported by Greg Linden, Brent Smith, and Jeremy York from Amazon.com in 2003 and were published on IEEE INTERNET COMPUTING. (http://www.cs.umd.edu/~samir/498/Amazon-Recommendations.pdf) In general, the main idea between these two algorithms is customers from similar background tends to make similar purchase choice. Thus, what the e-commence company did is to first learn from the previous customers’ shopping behavior and category him or her into a customer group. Customers within the same group share some similar shopping behavior. The online retailer also makes up the recommendation network based on their purchase history. Once these networks are established, whenever a customer come, the system will first figure out which group the customer belongs to and will give recommendation based on the group category.
    
    In this project, I will use the ABM model and network analysis to mimic the construction of this recommendation system and analyze how this system will impact the customers’ shopping behaviors.
    
    2.Agent Based Model and Network Analysis
    In this model, the construction of the recommendation system is an ABM problem. Every customer is served as an agent in this model. The products are environment and agents are make purchase decisions within the product environment. Whenever the agent makes up a purchase decision, a link between the two products will be given.  For each group of agents, since they are following different decision making process, their outcome which will be a recommendation network is expected to be different.  In this project, I will focus on how different decision making strategy will lead to different networks.
    
    On the other hand, another reason I choose ABM is the agent will affect the behavior of each other. The agent come into the network will be influenced by the agents before him, because the agent will tend to follow the existing recommendation system which will be determined by the previous purchase actions.
    
    Since the outcome of the ABM is a network, network analysis will be conducted to see how these networks are different from each other and to which extent they are different. 
    
    3.	The problem to answer
    a.	How different groups of customers will change the recommendation system, using eigenvector centrality as the evaluation method?
    b.	How many steps will it take to reach to an almost-complete recommendation system, using degree of centrality as the evaluation method?


##Methods
    1. Overview
    In this project, I will use the ABM to create the model and will use network analysis to analyze the outcome of the ABM which should be different networks. 
    
    During the ABM experiment, I will keep track of the decision making method (which is determined by the model parameters) and compare the eigenvector centrality and communicability of the outcome network. The eigenvector centrality will show the influence of each node and find out the most influential node and its corresponding variables. I will use these to compare networks. This will answer the first question listed above.
    
    I will also investigate the impact of each agent as well. Ideally, the agents come earlier are expected to have higher impact because the recommendation network is not completed and their purchase behaviors will work as the network build-up process. While on the other hand, if the agents enter the network later and the network is almost done, he or she will mostly follow the recommendation system and thus his or her behavior will have less impact on the whole network. For the network analysis, I will use the centrality algorithm to analyze how degree centrality of nodes. The degree centrality for a node v is the fraction of nodes it is connected to, thus when the degree centrality for each node is almost fix, we can assume that the network is almost fixed. During the ABM, I will keep track of the number of agents entering the model (for each step, a fixed number of customer will enter the system) and compare it to the degree of centrality and find out the completion point. This will answer the second question listed above.
    
    2. Details
    The most important process of this model is to conduct purchase by the customer. The purchase behavior will be conducted based on the Exploration-Exploitation mechanism given by James G. March (Exploration and Exploitation in Organizational Learning, Vol. 2, No. 1, Special Issue: Organizational Learning: Papers in Honor of (and by) James G. March (1991), pp. 71-87).  During the exploitation period, the decision strategy is to follow the best decision given current information, while during the exploration period, the decision strategy is to gather more information about the system thus it will choose not to follow the best decision given current information. In this model, the agents’ (customers) decision strategy will also follow this algorithm. The agent will get an epsilon value, which is the probability that the agent will follow the recommendation or not. The epsilon value will be a number close to 1 which means in the majority of time, the agent will follow the recommendation system, and this can ensure the performance and also be consistent with the real life activities.
    
    If the agent choose to follow recommendation system, then it will choose from the recommendation system based on its own preference. Basically, there are three preference rate (rank preference, (low) price preference and review preference). The agent will do a preference calculation:
    Preference = price*(1-price_preference)+rank*rank_preference+review*review_preference
    
    The price, rank and review are normalized to same mean. Since the price_preference means the preference for low price, 1-price_preference will be used in the calculation. This preference value will be used to decide which product the customer will choose from the recommendation system.
    
    Another important strategy in the decision making process is epsilon-greedy algorithm which is also a kind of Exploration-Exploitation. In this project, the agent will get a gamma value and it will have gamma probability to follow the product with highest preference value, while it will also keep a (1-gamma)*gamma probability to choose the second best, and so on.
    
    To sum up, when each agent enter the network, it will have a start product which is randomly generated and the start product number and choose from the recommendation list of this start product and decide which the next product should be. Then switch to the next product and repeat the process above until it reach the budget. Every step, a fixed number of customers will enter the network and for each step, the degree of centrality will be recorded. We assume that when the network is almost fixed, the degree of centrality of each node will be almost fixed as well.  The eigenvector centrality and the parameter of the node with the highest eigenvector centrality will also be recorded since we assume the node with the highest eigenvector centrality is the node with the highest influential power.

##Product Class

constructor: class constructor, which "initializes" or "creates" the product when we call Product().

Four methods get_id, get_price, get_rank, get_review will respectively return the price, rank, review and co_purchase product number. 

```python
class Product(object):
    """
    Product class, which encapsulates all the attributes of a product.
    """
    
    def __init__(self, model, product_id, product_price, product_rank, product_review):
        """
        Constructor for Product class.  By default,
          * it has a price generated from normal distribution
          * it has a rank generated from normal distribution
          * it has a review rate generated from normal distribution
        """
        # Set model link and ID
        self.model = model
        self.product_id = product_id
        
        # Set Product parameters.
        self.price = product_price
        self.rank = product_rank
        self.review = product_review
        
    def get_id(self):
        """
        Decide return the id of a product
        """
        return self.product_id
        
    def get_price(self):
        """
        Decide return the price of a product
        """
        return self.model.get_product_price(self.product_id)
    
    def get_rank(self):
        """
        Decide return the ranking of a product
        """
        return self.model.get_product_rank(self.product_id)
    
    def get_review(self):
        """
        Decide return the review of a product
        """
        return self.model.get_product_review(self.product_id)
```

#Customer Class
constructor: class constructor, which "initializes" or "creates" the customer when we call Customer().

Three methods get_id, get_budget, get_price_preference, get_rank_preference, get_review_preference, get_random_rate and get_priority_rate will return the budget, shopping preference, random purchase rate and priority rate of a customer.

```python
class Customer(object):
    """
     Customer class, which encapsulates all the attributes of a customer.
    """
    
    def __init__(self, model, customer_id, customer_price_preference, customer_review_preference, customer_rank_preference, customer_budget,
                 customer_random_rate, customer_priority_rate):
        """
        Constructor for Customer class.  By default,
          * it has a random rate which shows the possiblity that the customer will follow the recommendation or not, and this number will given by user
          * it has a priority rate which shows the possiblity that the customer follows the rank of the recommendation, and this number will given by user
          * it has a budget generated from normal distribution
          * it has rank, review and price preference rate given by the user
        """
        # Set model link and ID
        self.model = model
        self.customer_id = customer_id
        
        # Set Product parameters.
        self.price_preference = customer_price_preference
        self.rank_preference = customer_rank_preference
        self.review_preference = customer_review_preference
        self.budget = customer_budget
        self.random_rate = customer_random_rate
        self.priority_rate = customer_priority_rate
    
    def get_id(self):
        """
        Decide return the id of a customer
        """
        return self.customer_id
        
    def get_budget(self):
        """
        Decide return the budget of a customer
        """
        return self.model.get_customer_budget(self.customer_id)
    
    def get_rank_preference(self):
        """
        Decide return the rank preference of a customer
        """
        return self.model.get_customer_rank_preference(self.customer_id)
    
    def get_price_preference(self):
        """
        Decide return the price preference of a customer
        """
        return self.model.get_customer_price_preference(self.customer_id)
    
    def get_review_rank_preference(self):
        """
        Decide return the review preference of a customer
        """
        return self.model.get_customer_review_preference(self.customer_id)
    
    def get_purchase_rate(self):
        """
        Decide return the random purchase of a customer
        """
        return self.model.get_customer_random_rate(self.customer_id)
    
    def get_priority_rate(self):
        """
        Decide return the priority purchase rate of a customer
        """
        return self.model.get_customer_priority_rate(self.customer_id)
```

#Model Class
Below, we will define our model class. This can be broken up as follows:

constructor: class constructor, which "initializes" or "creates" the model when we call Model(). This is in the __init__ method.

setup_customer: method to create "customer"

setup_product: method to create "product"

setup_network: method to create our "network"

get_customer_price_preference, get_customer_rank_preference, get_customer_review_preference, get_customer_budget, get_customer_purchase_rate, get_product_price, get_product_rank, get_product_review, get_customer_random_rate, get_customer_priority_rate and get_co_purchase_product will return the customer's preference, budget, purchase rate, the product's price, ranking, review, random rate, priority rate and the co_purchase product list

purchase: This is a method simulating the purchase process of a single customer trigered by a start product. It will return a list of products a customer purchased through the network. The purchase will stop if the total amount of money exceeds the budget

step: main step method to control each time step simulation, in every step, a given number of  customers will enter the network and do purchase action

```python
class Model(object):
    numpy.random.seed(0)
    """
    Model class, which encapsulates the entire behavior of a single "run".
    """
    
    def __init__(self, product_number, customer_number, customer_budget_mean, customer_budget_std, customer_price_preference, 
                 customer_rank_preference, customer_review_preference, epsilon, gama,
                 product_price_mean, product_price_std, product_review_mean, product_review_std, product_rank_mean, product_rank_std):
        """
        Class constructor.
        """
        self.product_number = product_number    #The number of products involved in this model
        self.customer_number = customer_number    #The number of customers involved in this model
        
        #mean and std for customer budget
        self.customer_budget_mean = customer_budget_mean
        self.customer_budget_std = customer_budget_std
        
        #get the price, rank and review preference of the customer, these three variables will determine the customer group
        #these three variables will given by the test and serve as sweep parameters
        self.customer_price_preference = customer_price_preference
        self.customer_rank_preference = customer_rank_preference
        self.customer_review_preference = customer_review_preference
        
        #epsilon is the probability that the customer will follow the recommendation or not
        #epsilon is given by the test
        self.customer_random_rate = epsilon
        #gama is the probability that the customer will follow the best recommendation or not
        #gama is given by the test
        self.customer_priority_rate = gama        
        
        #mean and std for product price
        self.product_price_mean = product_price_mean
        self.product_price_std = product_price_std
        #mean and std for product review
        self.product_review_mean = product_review_mean
        self.product_review_std = product_review_std
        #mean and std for product rank
        self.product_rank_mean = product_rank_mean
        self.product_rank_std = product_rank_std
        
        # Set state variables
        self.product_co_purchase_list = []
        self.customer = []
        self.product = []
        
        # Setup history variables.
        self.history_network = []
        self.history_purchase = []
        self.history_graph = []
        self.history_centrality = []
        self.history_eigen_center_price = []
        self.history_eigen_center_rank = []
        self.history_eigen_center_review = []
        
        # Call setup methods to initialize customer, product, network
        self.setup_customer()
        self.setup_product()
        self.setup_network()
        
    def setup_customer(self):
        """
        Method to setup customer.
        """
        #set up customer list
        for i in xrange(self.customer_number):
            self.customer.append(Customer(model=self,
                                          customer_id=i,
                                          customer_price_preference = self.customer_price_preference,
                                          customer_rank_preference = self.customer_rank_preference, 
                                          customer_review_preference = self.customer_review_preference,
                                          customer_budget = numpy.random.normal(self.customer_budget_mean, self.customer_budget_std),
                                          customer_random_rate = self.customer_random_rate,
                                          customer_priority_rate = self.customer_priority_rate))
    
    def setup_product(self):
        """
        Method to setup all the products needed for the network.
        """
        #set up product list
        for i in xrange(self.product_number):
            self.product.append(Product(model=self,
                                      product_id=i,
                                      product_price=numpy.random.normal(self.product_price_mean,self.product_price_std),
                                      product_rank=numpy.random.normal(self.product_rank_mean,self.product_rank_std),
                                      product_review=numpy.random.normal(self.product_review_mean,self.product_review_std)))
    
    def setup_network(self):
        """
        Method to setup the co_purchase network
        """
        #for each customer, they will enter the network and do a purchase
        #first the customer will be assigned a start product and this start product is randomly generated
        #the customer will then start from this product and do purchase follow the decision strategy
        #the result of the customer purchase will be recorded and the co purchase pair will be store to create the network
        for customer in self.customer:
            start_product = numpy.random.randint(self.product_number, size=1)
            customer_purchase = self.purchase(customer, start_product)
            self.product_co_purchase_list.extend(customer_purchase)
        #create a network based on the current node(product) and the current network  
        product_array = numpy.arange(self.product_number)
        G=nx.Graph()
        G.add_nodes_from(product_array)
        for pair in self.product_co_purchase_list:
            G.add_edge(pair[0][0], pair[1][0])
        #calculate the degree of centrality of each node and get the sum
        centrality = nx.degree_centrality(G)
        total_centrality = sum(centrality.values())
        #calculate the eigenvector centrality of each node and find the most influential node
        try:
            eigen_centrality = nx.eigenvector_centrality(G)
            eigen_center = numpy.argmax(numpy.asarray(eigen_centrality.values()))
            self.history_eigen_center_price.append(self.product[eigen_center].get_price())
            self.history_eigen_center_rank.append(self.product[eigen_center].get_rank())
            self.history_eigen_center_review.append(self.product[eigen_center].get_review())
        except:
            self.history_eigen_center_price.append([None])
            self.history_eigen_center_rank.append([None])
            self.history_eigen_center_review.append([None])
        #store the network of each step, as well as the degree of centrality, eigenvector centrality
        self.history_centrality.append(total_centrality)
        self.history_graph.append(copy.deepcopy(G))
        self.history_network.append(self.product_co_purchase_list)
       

    def get_customer_review_preference(self, customer_id):
        """
        Get the review preference of a customer based on their ID.
        """
        
        return self.customer[customer_id].customer_review_preference
    
    def get_customer_price_preference(self, customer_id):
        """
        Get the price preference of a customer based on their ID.
        """
        
        return self.customer[customer_id].customer_price_preference
    
    def get_customer_rank_preference(self, customer_id):
        """
        Get the rank preference of a customer based on their ID.
        """
        
        return self.customer[customer_id].customer_rank_preference    
    
    def get_customer_budget(self, customer_id):
        """
        Get the budget of a customer based on their ID.
        """
        return self.customer[customer_id].budget

    
    def get_product_price(self, product_id):
        """
        Get the price of a product based on their ID.
        """
        return self.product[product_id].price
    
    
    def get_product_rank(self, product_id):
        """
        Get the rank of a product based on their ID.
        """
        return self.product[product_id].rank
    
    def get_product_review(self, product_id):
        """
        Get the review of a product based on their ID.
        """
        return self.product[product_id].review
    
       
    def get_co_purchase_product(self, product_id):
        """
        Get the co_purchase products of a product based on their ID.
        """
        co_purchase_list = []
        #To find all the co_purchase product, we need to search all the keys in the dictionary and find out all the co_purchase pairs
        for (product_1, product_2) in self.product_co_purchase_list:
            if product_1 == product_id:
                co_purchase_list.append(product_1)
            elif product_2 == product_id:
                co_purchase_list.append(product_2)
        return co_purchase_list
    
    def purchase(self, customer, start_product_id):
        """
        This is the basic step of the model. Once a purchase is done, it will change the network structure.
        A specific customer will be given to this model and the customer have a specific start product which he/she wants to buy initially
        The customer will then follow the decison strategy which is given by epsilon and gama and do purchase behavior through this method
        The output of this function is a list with product pairs the customer purchases
        """
        #get the customer's shopping parameters
        price_preference = customer.price_preference
        rank_preference = customer.rank_preference
        review_preference = customer.review_preference
        budget = customer.budget
        #get the customer's decison making parameters
        random_rate = customer.random_rate
        priority_rate = customer.priority_rate
        purchase_list = []
        #set up a parameter to track how much the customer spends
        total_money = self.product[start_product_id].get_price()
        #assign the initail product
        customer_select_product = start_product_id
        while True:
            #if the total amount exceeds the budget, the customer purchase stops
            #The purchase list will be stored in the history_purchase list
            if total_money >= budget:
                self.history_purchase.append(purchase_list)
                return purchase_list
                break
            #if a customer decide to buy a product, then we will look at the all the co_purchase products around.
            #The customer has two options: follow the recommendation or not
            #if not follow, it will randomly pick a product from the network
            #if follow, I will calculate a preference score for each co_purchase products based on the customer's shopping preference
            #the customer will then choose from the preference and follow the epsilon greedy policy with probability gamma            
            whether_random = numpy.random.binomial(1, random_rate)
            if whether_random == 1:
                purchase_id = numpy.random.randint(self.product_number, size=1)
            else:
                co_purchase_list = self.get_co_purchase_product(customer_select_product)
                
                if co_purchase_list:
                    purchase_preference = []
                    for i in xrange(len(co_purchase_list)):
                        price = self.get_product_price(co_purchase_list[i])
                        rank = self.get_product_rank(co_purchase_list[i])
                        review = self.get_product_review(co_purchase_list[i])
                        preference = price*(1-price_preference)+rank*rank_preference+review*review_preference
                        purchase_preference.append(preference)
                    preference_order = numpy.argsort(numpy.asarray(purchase_preference)).tolist()
                    for i in xrange(len(preference_order)):
                        prob = numpy.random.binomial(1, priority_rate)
                        if prob == 1:
                            loc = preference_order.index(i)
                            purchase_id = co_purchase_list[loc]
                            break
                        if i == len(preference_order)-1:
                            purchase_id = numpy.random.randint(self.product_number, size=1)
                else:
                    purchase_id = numpy.random.randint(self.product_number, size=1)
            purchase_list.append((customer_select_product, purchase_id))
            purchase_price = self.product[purchase_id].get_price()
            total_money += purchase_price                           
            customer_select_product = purchase_id
    
    def step(self, customers):
        """
        *For each step, a group of customers will enter the network and do purchase
        *For each step, the network will be updated since more purchases are made and more edges will be added
        *For each step, the total degree of centrality and the eigenvector centrality will be calculated.
        *For each step, the total degree of centrality and the rank, review and price of the node with the highest eigenvector centrality will be recorded
        """
        for customer in customers:
            start_product = numpy.random.randint(self.product_number, size=1)
            customer_purchase = self.purchase(customer, start_product)
            self.product_co_purchase_list.extend(customer_purchase) 
        #create a network based on the current node(product) and the current network  
        product_array = numpy.arange(self.product_number)
        G=nx.Graph()
        G.add_nodes_from(product_array)
        for pair in self.product_co_purchase_list:
            G.add_edge(pair[0][0], pair[1][0])
        #calculate the degree of centrality of each node and get the sum
        centrality = nx.degree_centrality(G)
        total_centrality = sum(centrality.values())
        #calculate the eigenvector centrality of each node and find the most influential node
        eigen_centrality = nx.eigenvector_centrality(G)
        eigen_center = numpy.argmax(numpy.asarray(eigen_centrality.values()))
        #store the network of each step, as well as the degree of centrality, eigenvector centrality
        self.history_centrality.append(total_centrality)
        self.history_graph.append(copy.deepcopy(G))
        self.history_network.append(self.product_co_purchase_list)
        self.history_eigen_center_price.append(self.product[eigen_center].get_price())
        self.history_eigen_center_rank.append(self.product[eigen_center].get_rank())
        self.history_eigen_center_review.append(self.product[eigen_center].get_review())
            
```

```python
# Set number of samples per value and steps per sample
num_samples = 200
num_steps = 10

# Set basic model parameters
product_number = 10
customer_number = 10
customer_budget_mean=35
customer_budget_std=10
epsilon=0.9
gama=0.9
product_price_mean=10
product_price_std=5
product_review_mean=10
product_review_std=2
product_rank_mean=10
product_rank_std=2


# Set subsidy values to "sweep" over
preference_sweep_values = [[0.9,0.05,0.05], [0.1,0.45,0.45], [0.33,0.33,0.33]]
sweep_centrality = []
sweep_graph = []
sweep_eigen_price = []
sweep_eigen_rank = []
sweep_eigen_review = []
        
for i in xrange(3):
    sweep_centrality.append([])
    sweep_graph.append([])
    sweep_eigen_price.append([])
    sweep_eigen_rank.append([])
    sweep_eigen_review.append([])


# Iterate over values
for i in xrange(3):
    preference = preference_sweep_values[i]
    for j in xrange(num_samples):
        # Output info
        m = Model(product_number=product_number, customer_number=customer_number, 
                  customer_budget_mean=customer_budget_mean, customer_budget_std=customer_budget_std, 
                  customer_price_preference=preference[0], customer_rank_preference=preference[1], 
                  customer_review_preference=preference[2], epsilon=epsilon, gama=gama,    
                  product_price_mean=product_price_mean, product_price_std=product_price_std, 
                  product_review_mean=product_review_mean, product_review_std=product_review_std, 
                  product_rank_mean=product_rank_mean, product_rank_std=product_rank_std)
        temp = []
        for t in xrange(num_steps):
            m.step(m.customer)
        sweep_centrality[i].append(m.history_centrality)
        sweep_graph[i].append(m.history_graph)
        sweep_eigen_price[i].append(m.history_eigen_center_price)
        sweep_eigen_rank[i].append(m.history_eigen_center_rank)
        sweep_eigen_review[i].append(m.history_eigen_center_review)       
```

    1. How customer number affect the network's degree of centrality.
    In this experiment, I examined how the number of customers will affect the degree of centrality of the network. From the graph below, we can see that all three groups have the same trend. At the beginning of the test, the network is almost empty and every ten customers will make a huge difference to the networks and the total degree of centrality increases a lot at the beginning. However, as the customer number increases, the increment decreases, which means the majority of the customers are following the recommendation system and the system is well-established, thus fewer new edges are added. We can see for a simple product group of 10 product, about 50 customers can almost establish the recommendation network and this does not depend on what customer group they are belonging to.
    
    This experiment is to answer how the number of the customers will affect the network and how many steps the network need to reach a stable condition. We can see that the customers come in earlier are served as the constructor and their behavior will largely change the structure and the customers come later will have less influence. Furthermore, there are about 50 customers needed for a ten product network to reach a relatively stable condition. Since the decision strategy used in this project is exploration-exploitation and epsilon greedy, it should converge to a stable condition.

However, when the number of customers increase, the prices on the center nodes come close to each other. In other word, a large number of customers enter the network, the customers’ preference will have little influence on the network.

However, from the second graph and the third graph, we can see that the performance of the center node rank and review is almost random. One explanation is that since price counts the largest weight in preference calculation, the rank and review factors do not influence the choice decision too much.
